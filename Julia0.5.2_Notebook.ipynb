```julia
using QuantEcon: meshgrid

n = 50
x = linspace(-3, 3, n)
y = x

z = Array{Float64}(n, n)
f(x, y) = cos(x^2 + y^2) / (1 + x^2 + y^2)
for i in 1:n
    for j in 1:n
        z[j, i] = f(x[i], y[j])
    end
end

fig = figure(figsize=(8,6))
ax = fig[:gca](projection="3d")
ax[:set_zlim](-0.5, 1.0)
xgrid, ygrid = meshgrid(x, y)
ax[:plot_surface](xgrid, ygrid, z, rstride=2, cstride=2,
  cmap=ColorMap("jet"), alpha=0.7, linewidth=0.25)
```

```julia
using PyPlot

## global variables
const L = 20;
const N = L^2;
const m = 10^5;
const burnin_m = 10^3;
const T_crit = 2.0/log(1.0+sqrt(2));
const Tmin = 1;
const Tmax = 6;
const N_iter = 40;
```

```julia
## given a site index, return nearest neighbors
function nneighbor(x::Int64)
    (i, j) = (div(x,L), x%L)
    
    ## up
    if j != 0
        xn_u = i*L + (j-1)
    else
        xn_u = i*L + (L-1)
    end
    
    ## down
    if j != L-1
        xn_d = i*L + (j+1)
    else
        xn_d = i*L + 0
    end
    
    ## left
    if i != 0
        xn_l = (i-1)*L + j
    else
        xn_l = (L-1)*L + j
    end
    
    ## right
    if i != L-1
        xn_r = (i+1)*L + j
    else
        xn_r = 0*L + j
    end
    
    return [xn_u, xn_d, xn_l, xn_r]
                end


## energy
function energy(lattice::Array{Int64,1})
    en = 0
    for x in 0:N-1
        for x_nn in nneighbor(x)
            en += -0.5*lattice[x_nn+1]*lattice[x+1]
        end
    end
    return en/N
end


## metropolis algorithm: preform 1 sweep over the N spins
function metropolis(lattice::Array{Int64,1}, temp::Float64)
    beta = 1.0/temp
    random = rand(N) #the random number involved in the monte carlo
    random_spin = rand(0:N-1, N) #the random spin
    for i in 0:N-1        
        x = random_spin[i+1]
        E_spin = 0
        delta_E = 0
        for x_nn in nneighbor(x)
            E_spin += -lattice[x+1]*lattice[x_nn+1]
        delta_E = -2*E_spin
        end
        
        if delta_E <= 0
            lattice[x+1] = (-1)*lattice[x+1]
        else
            prob = exp(-beta*delta_E)
            if prob > random[i+1]
                lattice[x+1] = (-1)*lattice[x+1]
            end
        end 
    end
    return lattice
end
```

```julia
## inspect a critical sample
tic();
crit_lattice = 2*rand(0:1, N) - 1
for i in 1:10^3
    crit_lattice = metropolis(crit_lattice, T_crit)
end

toc()

println(energy(crit_lattice))
matshow(reshape(crit_lattice, L, L))
```

```julia
tic();

## Compute thermodynamic quantities as a function of temperature
T_list = collect(linspace(Tmin, Tmax, N_iter));
energy_list = zeros(N_iter)
energy_sq_list = zeros(N_iter)
mag_list = zeros(N_iter)

for i in 1:N_iter
    
    tmp_energy_list = zeros(m - burnin_m)
    tmp_energy_sq_list = zeros(m - burnin_m)
    tmp_mag_list = zeros(m - burnin_m)

    lattice = 2*rand(0:1, N) - 1
    for j in 1:m
        lattice = metropolis(lattice, T_list[i])
        if j > burnin_m
            tmp_energy_list[j - burnin_m] = energy(lattice)
            tmp_energy_sq_list[j- burnin_m] = energy(lattice)^2
            tmp_mag_list[j - burnin_m] = sum(lattice)/N
        end
    end
    energy_list[i] = sum(tmp_energy_list)/(m - burnin_m)
    energy_sq_list[i] = sum(tmp_energy_sq_list)/(m - burnin_m)
    mag_list[i] = sum(tmp_mag_list)/(m - burnin_m)
end

toc()
```

```julia
plot(T_list, energy_list, color="red", linewidth=2.0, linestyle="--")
```

```julia
plot(T_list, (energy_sq_list - energy_list.^2).*(T_list.^(-2)), color="red", linewidth=2.0, linestyle="--")
```

```julia
plot(T_list, abs(mag_list), color="red", linewidth=2.0, linestyle="--")
```
